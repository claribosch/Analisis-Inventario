plt.pie: Se utiliza cuando se desea crear un único gráfico de torta en la figura actual sin subplots explícitos.
ax1.pie: Se utiliza cuando se crean subplots con plt.subplots y se quiere dibujar en un eje específico. Esto permite tener múltiples gráficos en la misma figura, cada uno con su propio conjunto de datos y propiedades.

El uso de ax1.pie (o cualquier otro método de ax) es necesario para controlar de manera independiente múltiples subplots en una sola figura, lo que proporciona mayor flexibilidad y organización en la visualización de datos.

cuando ejecutas plt.subplots(), matplotlib crea instancias de las clases Figure y Axes y te las devuelve. Vamos a profundizar más en esto.

Proceso de plt.subplots()
Creación de la Figura y los Ejes:

plt.subplots() crea una figura (Figure) y una cuadrícula de subplots (Axes).
La función devuelve una tupla que contiene la figura y una matriz de objetos Axes.
Instancias de Clases:

fig es una instancia de la clase Figure.
ax (o ax1, ax2, etc.) son instancias de la clase Axes

ax.pie es una función que pertenece al objeto ax, que es una instancia de la clase Axes en matplotlib. En términos de programación orientada a objetos, ax es un objeto que tiene métodos y propiedades que se pueden utilizar para manipular y personalizar los gráficos.

Desglose en Términos de Programación
Objeto ax:

ax es una instancia de la clase Axes en matplotlib.
Un objeto Axes representa un solo gráfico en una figura.
Tiene métodos para crear diferentes tipos de gráficos (como pie, plot, bar, etc.) y para personalizar aspectos del gráfico (como títulos, etiquetas, límites de ejes, etc.).
Método ax.pie:

ax.pie es un método del objeto Axes que se utiliza para dibujar un gráfico de torta en ese eje específico.
Cuando llamas a ax.pie(...), estás utilizando un método que pertenece al objeto ax para crear y personalizar un gráfico de torta.
