plt.pie: Se utiliza cuando se desea crear un √∫nico gr√°fico de torta en la figura actual sin subplots expl√≠citos.
ax1.pie: Se utiliza cuando se crean subplots con plt.subplots y se quiere dibujar en un eje espec√≠fico. Esto permite tener m√∫ltiples gr√°ficos en la misma figura, cada uno con su propio conjunto de datos y propiedades.

El uso de ax1.pie (o cualquier otro m√©todo de ax) es necesario para controlar de manera independiente m√∫ltiples subplots en una sola figura, lo que proporciona mayor flexibilidad y organizaci√≥n en la visualizaci√≥n de datos.

cuando ejecutas plt.subplots(), matplotlib crea instancias de las clases Figure y Axes y te las devuelve. Vamos a profundizar m√°s en esto.

Proceso de plt.subplots()
Creaci√≥n de la Figura y los Ejes:

plt.subplots() crea una figura (Figure) y una cuadr√≠cula de subplots (Axes).
La funci√≥n devuelve una tupla que contiene la figura y una matriz de objetos Axes.
Instancias de Clases:

fig es una instancia de la clase Figure.
ax (o ax1, ax2, etc.) son instancias de la clase Axes

ax.pie es una funci√≥n que pertenece al objeto ax, que es una instancia de la clase Axes en matplotlib. En t√©rminos de programaci√≥n orientada a objetos, ax es un objeto que tiene m√©todos y propiedades que se pueden utilizar para manipular y personalizar los gr√°ficos.

Desglose en T√©rminos de Programaci√≥n
Objeto ax:

ax es una instancia de la clase Axes en matplotlib.
Un objeto Axes representa un solo gr√°fico en una figura.
Tiene m√©todos para crear diferentes tipos de gr√°ficos (como pie, plot, bar, etc.) y para personalizar aspectos del gr√°fico (como t√≠tulos, etiquetas, l√≠mites de ejes, etc.).
M√©todo ax.pie:

ax.pie es un m√©todo del objeto Axes que se utiliza para dibujar un gr√°fico de torta en ese eje espec√≠fico.
Cuando llamas a ax.pie(...), est√°s utilizando un m√©todo que pertenece al objeto ax para crear y personalizar un gr√°fico de torta.
